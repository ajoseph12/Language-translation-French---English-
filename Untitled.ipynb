```python
sentences = []
with open("OpenSubtitles.en-fr.en", 'r') as reader:
        for s in reader:
            sentences.append(s.strip())
```

```python
sentences
```

```python
from collections import Counter
```

```python
en_vocab_dict = Counter(word.strip(',." ;:)(][?!') for sentence in sentences for word in sentence.split())
```

```python
en_vocab = list(map(lambda x: x[0], sorted(en_vocab_dict.items(), key = lambda x: -x[1])))
```
